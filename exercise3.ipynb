Pre-defined Circular Queue class

```python
class CircularQueue():

    def __init__(self, k):
        self.k = k
        self.queue = [None] * k
        self.head = self.tail = -1

    # Insert an element into the circular queue
    def enqueue(self, data):

        if ((self.tail + 1) % self.k == self.head):
            print("The circular queue is full\n")

        elif (self.head == -1):
            self.head = 0
            self.tail = 0
            self.queue[self.tail] = data
        else:
            self.tail = (self.tail + 1) % self.k
            self.queue[self.tail] = data

    # Delete an element from the circular queue
    def dequeue(self):
        if (self.head == -1):
            print("The circular queue is empty\n")

        elif (self.head == self.tail):
            temp = self.queue[self.head]
            self.head = -1
            self.tail = -1
            return temp
        else:
            temp = self.queue[self.head]
            self.head = (self.head + 1) % self.k
            return temp

    def printCQueue(self):
        if(self.head == -1):
            print("No element in the circular queue")

        elif (self.tail >= self.head):
            for i in range(self.head, self.tail + 1):
                print(self.queue[i], end=" ")
            print()
        else:
            for i in range(self.head, self.k):
                print(self.queue[i], end=" ")
            for i in range(0, self.tail + 1):
                print(self.queue[i], end=" ")
            print()
```

#### 2. Create a Deque

Given the queue below,

**Queue:**

| Cat | Dog | Rabbit | Bird | Hamster
| --- | --- | --- | --- | --- |

**dequeue** each element and insert them into a **Deque** in the following order below:

**Deque:**

| Hamster | Bird | Rabbit | Dog | Cat
| --- | --- | --- | --- | --- |

**Expected output:**

['Hamster','Bird', 'Rabbit', 'Dog', 'Cat']

```python
#DEQUE

class Deque:
  def __init__(self):
    self.animals = ["Cat", "Dog", "Rabbit", "Bird", "Hamster"]

  def display(self):
    print(self.animals)

  def removeRear(self): #pop each out
    return self.animals.pop()

  def addFront(self, item): #position it
    self.animals.insert(0, item)

  def newDeque(self): #deque location
    animals_deque = []
    while self.animals:
      item = self.removeRear()
      animals_deque.append(item)

    print(animals_deque)

animals = Deque() #create instace

print("Initial Position")
animals.display() #first position

print("After Deque: ")
animals.newDeque() #new position
```

#### 3. Create a Circular Queue

Given the queue below,

**Queue:**

| I | II | III | IV | V | VI | VII | VIII | IX | X | XI | XII
| --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- |

**dequeue** each element and insert them into a **Circular Deque** in the following order below:

**Circular Queue:**

<img src="clock.jpg"/>

**Expected output:**

I II III IV V VI VII VIII IX X XI XII

```python
# Circular Queue

class CircularQueue:
    def __init__(self, k):
        self.k = k
        self.queue = [None] * k
        self.front = self.rear = -1

    def is_empty(self):
        return self.front == -1

    def is_full(self):
        return (self.rear + 1) % self.k == self.front

    def enqueue(self, element):
        if self.is_full():
            print("Queue is full")
        else:
            if self.is_empty():
                self.front = 0
            self.rear = (self.rear + 1) % self.k
            self.queue[self.rear] = element

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty")
            return None
        else:
            element = self.queue[self.front]
            if self.front == self.rear:
                self.front = -1
                self.rear = -1
            else:
                self.front = (self.front + 1) % self.k
            return element

# define elems
elements = ["I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX", "X", "XI", "XII"]

# define k and create instance
k = len(elements)
circular_queue = CircularQueue(k)

# deque elem and put it to circular queue
for element in elements:
    circular_queue.enqueue(element)

# display elem (imagine it to be in circular order, cant do the image)
while not circular_queue.is_empty():
    print(circular_queue.dequeue())
```

#### 4. Create a Priority Queue

Given the queue below,

**Queue:**

| Cat | Dog | Rabbit | Bird | Hamster
| --- | --- | --- | --- | --- |


**dequeue** each element and insert them into a **Priority Queue** with the following priority values

**Priority Queue:**

| Element | Priority |
| --- | --- |
| Hamster |  0 |
| Bird |  1 |
| Rabbit |  2 |
| Dog |  3 |
| Cat |  4 |

**Expected output:**

Hamster Bird Rabbit Dog Cat

```python
# Priority Queue

class PriorityQueue:
    def __init__(self):
        self.elements = []

    def push(self, item, priority):
        self.elements.append((item, priority))

    def pop(self):
        return self.elements.pop(0)[0]

    def is_empty(self):
        return len(self.elements) == 0

# assing elem and values
queue_elements = ["Cat", "Dog", "Rabbit", "Bird", "Hamster"]
values = [4, 3, 2, 1, 0]

# create instance
priority_queue = PriorityQueue()
#push elem
for i in range(len(queue_elements)):
    element = queue_elements[i]
    priority = values[i]
    priority_queue.push(element, priority)

# sort each elem based on priority value
priority_queue.elements.sort(key=lambda x: x[1])

# pop elem
priority_order = []
while not priority_queue.is_empty():
    priority_order.append(priority_queue.pop())

print("Priority Queue:", priority_order)

```
